```python
import matplotlib.pyplot as plt
import matplotlib.colors as colors
plt.style.use('belle2')
import numpy as np
import pdg
import root_pandas
import pandas

filename='../Ntuples/N_e_mass1_width1_2.root'
df = root_pandas.read_root(filename,key='B0')

df=df.drop_duplicates(subset=['__experiment__','__run__','__event__','__candidate__'])
df=df.dropna(subset=['B0_vtxReChi2','D_BFM'])
#df = df.query('nElectrons90==3')
df['B0_mcPDG'] = df['B0_mcPDG'].fillna(0)

df.eval('B_D_ReChi2 = B0_vtxReChi2 + D_vtxReChi2', inplace=True)
df.eval('q2_SigOnly = (0.5*Ecms - D_CMS_E)**2 - D_CMS_px**2 - D_CMS_py**2 - D_CMS_pz**2', inplace=True)
```

```python
def M_ee_gen(e2):
    # Energy
    E = df.e_1_mcE + df[e2+'_mcE']
    # dot product
    p_xdot = df.e_1_mcPX * df[e2+'_mcPX']
    p_ydot = df.e_1_mcPY * df[e2+'_mcPY']
    p_zdot = df.e_1_mcPZ * df[e2+'_mcPZ']
    p_dot = p_xdot + p_ydot + p_zdot
    # Calculating the final quantities
    M_ee2 = E**2 - df.e_1_mcP**2 - df[e2+'_mcP']**2 - 2*p_dot
    M_ee = M_ee2**0.5
    return M_ee


def M_ee(e2):
    # Energy
    E = df.e_1_E + df[e2+'_E']
    # dot product
    p_xdot = df.e_1_px * df[e2+'_px']
    p_ydot = df.e_1_py * df[e2+'_py']
    p_zdot = df.e_1_pz * df[e2+'_pz']
    p_dot = p_xdot + p_ydot + p_zdot
    # Calculating the final quantities
    M_ee2 = E**2 - df.e_1_p**2 - df[e2+'_p']**2 - 2*p_dot
    M_ee = M_ee2**0.5
    return M_ee


def opening_angle_gen(obj1, obj2):
    # dot product
    p_xdot = df[obj1+'_mcPX'] * df[obj2+'_mcPX']
    p_ydot = df[obj1+'_mcPY'] * df[obj2+'_mcPY']
    p_zdot = df[obj1+'_mcPZ'] * df[obj2+'_mcPZ']
    p_dot = p_xdot + p_ydot + p_zdot
    # cosin
    cosin = p_dot / (df[obj1+'_mcP'] * df[obj2+'_mcP'])
    return cosin


def opening_angle(obj1, obj2):
    # dot product
    p_xdot = df[obj1+'_px'] * df[obj2+'_px']
    p_ydot = df[obj1+'_py'] * df[obj2+'_py']
    p_zdot = df[obj1+'_pz'] * df[obj2+'_pz']
    p_dot = p_xdot + p_ydot + p_zdot
    # cosin
    cosin = p_dot / (df[obj1+'_p'] * df[obj2+'_p'])
    return cosin


def angular_resolution(e2,angle):
    # Generator level angles
    theta_N_gen = df.e_1_mcTheta - df[e2+'_mcTheta']
    phi_N_gen = df.e_1_mcPhi - df[e2+'_mcPhi']
    # Reconstructed angles
    theta_N_reco = df.e_1_theta - df[e2+'_theta']
    phi_N_reco = df.e_1_phi - df[e2+'_phi']
    # Difference
    theta_N_resolution = theta_N_gen - theta_N_reco
    phi_N_resolution = phi_N_gen - phi_N_reco
    if angle=='theta':
        return theta_N_resolution
    if angle=='phi':
        return phi_N_resolution
    else:
        print('Error!')


# assume the opening angle between the e+e- from N is bigger
condition = (opening_angle('e_0','e_1')>=opening_angle('e_2','e_1')) & (df.e_0_charge==-df.e_1_charge)
df['e_B_genMotherPDG'] = np.where(condition, df['e_0_genMotherPDG'],df['e_2_genMotherPDG'])

df['M_ee_gen'] = np.where(condition, M_ee_gen('e_0'),M_ee('e_2'))
df['M_ee'] = np.where(condition, M_ee('e_0'),M_ee('e_2'))
df['M_ee_resolution'] = df['M_ee_gen'] - df['M_ee']

df['angle_ee_gen'] = np.where(condition,opening_angle_gen('e_0','e_1'),opening_angle_gen('e_2','e_1'))
df['angle_ee_gen'] = np.arccos(df['angle_ee_gen'])

df['angle_ee'] = np.where(condition,opening_angle('e_0','e_1'),opening_angle('e_2','e_1'))
df['angle_ee'] = np.arccos(df['angle_ee'])
df['angle_ee_resolution'] = df['angle_ee_gen'] - df['angle_ee']

df['theta_ee_resolution'] = np.where(condition,angular_resolution('e_0','theta'),angular_resolution('e_2','theta'))
df['phi_ee_resolution'] = np.where(condition,angular_resolution('e_0','phi'),angular_resolution('e_2','phi'))

std1 = 0.0018
std2 = 0.0023
df['theta_chi']=np.abs(df['theta_ee_resolution']) / std1
df['phi_chi']=np.abs(df['phi_ee_resolution']) / std2


df['e_B_genMotherPDG'] = np.where(condition, df['e_2_genMotherPDG'],df['e_0_genMotherPDG'])
df['e_B_px'] = np.where(condition, df['e_2_px'],df['e_0_px'])
df['e_B_py'] = np.where(condition, df['e_2_py'],df['e_0_py'])
df['e_B_pz'] = np.where(condition, df['e_2_pz'],df['e_0_pz'])
df['e_B_pt'] = np.where(condition, df['e_2_pt'],df['e_0_pt'])
df['e_B_p'] = np.where(condition, df['e_2_p'],df['e_0_p'])

df.eval('p_Y = sqrt( (D_px + e_B_px)**2 + (D_py + e_B_py)**2 +(D_pz + e_B_pz)**2 )', inplace=True)

# Best Candidate selection
df_bestSelected=df.loc[df.groupby(['__experiment__','__run__','__event__','__production__']).B_D_ReChi2.idxmin()]
#df=df.query('e_0_pt>0.2 and e_1_pt>0.2 and e_2_pt>0.2')
```

```python
import scipy
import scipy.stats

def signal_pdf(x,mass,width,normalization):
    return normalization * scipy.stats.norm.pdf(x, loc=mass, scale=width)

def logLikelihood(x,parameters):
    mass,width,normalization = parameters
    return -np.log(signal_pdf(mass, width, normalization)).sum()

result = scipy.optimize.minimize(logLikelihood, [0, 0.02, 600], args=(signal_bestSelected.query(cut)[variable]), method='Nelder-Mead')
print(f"Fit success: {result.success}, mass: {result.x[0]}, width: {result.x[1]}")
```
